```python
import seaborn as sns
import numpy as np
import pandas as pd
```

# 绘制单变量分布

```python
np.random.seed(0) # 确定随机数⽣成器的种⼦,如果不使⽤每次⽣成图形不⼀样
arr = np.random.randn(100)
sns.distplot(arr, bins=10, hist=True, kde=True, rug=True)
```

# 绘制双变量分布
## 绘制散点图

```python
df = pd.DataFrame({'x':np.random.randn(500), 'y':np.random.randn(500)})
df.head()
```

```python
sns.jointplot(x='x', y='y', data=df, kind='scatter', color='r', size=10, ratio=5, space=1)
```

## 绘制⼆维直⽅图

```python
sns.jointplot(x="x", y="y", data=df, kind='hex')
```

## 绘制核密度估计图形

```python
sns.jointplot(x="x", y="y", data=df, kind='kde')
```

# 类别散点图

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.stripplot(x='day', y='total_bill', data = data, hue='time', jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data = data, hue='time', jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data = data)
```


# 类别内的数据分布
## 箱形图

```python
sns.boxplot(x='day', y='total_bill', data=data, hue='time', palette=['g', 'r'], saturation=0.5)
```

## 绘制提琴图

```python
sns.violinplot(x='day', y='total_bill', data=data)
```

# 类别内的统计估计

```python
sns.barplot(x='day', y='total_bill', data=data)
```

```python
sns.pointplot(x='day', y='total_bill', data=data)
```
